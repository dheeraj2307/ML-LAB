```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
data = pd.read_csv('College.csv')
data.head()
```

```python
data = data.drop(['College'],axis=1)
from sklearn.preprocessing import LabelEncoder
lb_private = LabelEncoder()
data['Private'] = lb_private.fit_transform(data['Private'])
data.head()
```

```python
corr = data.corr()
corr
```

```python
ind = np.where(corr>0.7)
ind = np.array(ind)
del_ind = []
for i in range(0,34):
    if ind[0,i]>ind[1,i]:
        del_ind.append(ind[1,i])
del_ind = np.unique(np.array(del_ind))
print(del_ind)
data = np.array(data)
data = np.delete(data,del_ind,axis=1)
data = data.astype(int)
```

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte = train_test_split(data[:,1:],data[:,0])
data
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
kmeans = KMeans(n_clusters = 2)
kmeans.fit(xtr)
```

```python
y_pred = kmeans.predict(xte)
# print(np.concatenate((y_pred.reshape(len(y_pred),1),yte.reshape(len(y_pred),1)),axis=1))
cf = confusion_matrix(y_pred,yte)
cf
```

```python
precision = (cf[0][0]/(cf[0][0]+cf[0][1]))
recall = (cf[0][0]/(cf[0][0]+cf[1][0]))
f1 = 2*precision*recall/(precision+recall)
accuracy = 100*(cf[0][0]+cf[1][1])/(cf[0][0]+cf[1][1]+cf[0][1]+cf[1][0])
```

```python
print("Accuracy is: %.2f\nPrecision is: %.2f\nRecall is: %.2f\nF1_Score is: %.2f"%(accuracy,precision,recall,f1))
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters = 2)  #Hierarchical Clustering
agg.fit(xtr)
```

```python
y_pred = agg.fit_predict(xte)
# print(np.concatenate((y_pred.reshape(len(y_pred),1),yte.reshape(len(y_pred),1)),axis=1))
cf = confusion_matrix(y_pred,yte)
cf
```

```python
precision = (cf[0][0]/(cf[0][0]+cf[0][1]))
recall = (cf[0][0]/(cf[0][0]+cf[1][0]))
f1 = 2*precision*recall/(precision+recall)
accuracy = 100*(cf[0][0]+cf[1][1])/(cf[0][0]+cf[1][1]+cf[0][1]+cf[1][0])
print("Accuracy is: %.2f\nPrecision is: %.2f\nRecall is: %.2f\nF1_Score is: %.2f"%(accuracy,precision,recall,f1))
```
